```python
#%%
import time

import keras.optimizers
import pydot
from keras.models import Sequential
from keras.layers import Dense, Flatten, Reshape
from keras.layers import LSTM, Conv1D, MaxPooling1D
import pickle
from numpy import sqrt
from sklearn.metrics import mean_squared_error
import pydot
import pydotplus
import graphviz
from keras.utils import plot_model
from keras_visualizer import visualizer
```

```python
path_cnn = 'standardised_1DCNN_TS_forecasting_10ts_9nfeatures_9units.h5'
cnn = keras.models.load_model(path_cnn)
```

```python
path_lstm = 'standardised_LSTM_TS_forecasting_10ts_9nfeatures_2layers_40units.h5'
lstm = keras.models.load_model(path_lstm)
```

```python
cnn.summary()
```

```python
lstm.summary()
```

```python
visualizer(cnn, file_name='cnn_graph', file_format='png', view=True, settings=None)
```

```python
visualizer(lstm, file_name='lstm_graph', file_format='png', view=True, settings=None)
```

```python
plot_model(cnn, to_file='cnn_Stucrture_plot.png',
                          show_shapes=True,
                          show_layer_activations=True,
                          show_dtype=True,
                          show_layer_names=True)
```

```python
plot_model(lstm, to_file='LSTM_Stucrture_plot.png',
                          show_shapes=True,
                          show_layer_activations=True,
                          show_dtype=True,
                          show_layer_names=True)
```
